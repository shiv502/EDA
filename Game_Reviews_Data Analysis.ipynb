```python
# 'Boolean indexing'
#As we saw above, the mean of all the values in the score column of reviews is around 7. 
#What if we wanted to find all the games that got an above average score i.e, >7...?

above_avg_score = movie_reviews['score']>7
above_avg_score
```

```python
# Once we have a Boolean Series, we can use it to select the rows whose values are True 
#So, we could only select rows in reviews where score is greater than 7:

score_true_reviews = movie_reviews[above_avg_score]
score_true_reviews
```

```python
# It's possible to use multiple conditions for filtering. Let's say we want to find games released 
# for Xbox One that have a score of more than 7.

Xbox_one_filter = (movie_reviews['score']>7) & (movie_reviews['platform']== 'Xbox One')
filtered_reviews = movie_reviews[Xbox_one_filter]
filtered_reviews.head(10)
```
